```python
test1=movie.merge(credits,on='title')

```

```python
test1=test1[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
test1.info()
```

```python
test1.isnull().sum()
test1.dropna(inplace=True)
```

```python
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
import ast 
```

```python
test1['genres']=test1['genres'].apply(convert)
```

```python
test1['keywords']=test1['keywords'].apply(convert)
```

```python
def convert2(obj):
  count=0
  l=[]
  for i in ast.literal_eval(obj):
    if count!=3:
      l.append(i['name'])
      count=count+1
    else:
      break
  return l
```

```python
test1['cast']=test1['cast'].apply(convert2)
```

```python
def fetch_dir(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
test1['crew']=test1['crew'].apply(fetch_dir)
```

```python
test1.head()
```

```python
test1['overview']=test1['overview'].apply(lambda x:x.split())
```

```python
test1.head(1)
```

```python
test1['keywords']=test1['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
test1['cast']=test1['cast'].apply(lambda x:[i.replace(" ","") for i in x])
test1['crew']=test1['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
test1.head()
```

```python
test1['tags']=test1['overview']+test1['genres']+test1['keywords']+test1['cast']+test1['crew']
```

```python
test1['tags']
```

```python
df= test1[['movie_id','title','tags']]
```

```python
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors= cv.fit_transform(df['tags']).toarray()
```

```python
len(cv.get_feature_names())
```

```python
import nltk
```

```python
from  nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i  in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar= cosine_similarity(vectors)
similar
```

```python
def recommend(m):
  movie_index= df[df['title']==m].index[0]
  distances= similar[movie_index]
  movie_list= sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movie_list:
    print(df.iloc[i[0]].title)
  
```
